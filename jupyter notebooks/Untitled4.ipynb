```python
from bs4 import BeautifulSoup
import requests
import json
```

```python
url = "https://fantasy.premierleague.com/api/"
all_suffix = "bootstrap-static/"

content_ = requests.get(url+all_suffix).content
```

```python
json_all = json.loads(content_)
```

```python
json_all['events']
```

```python
def get_current_gameweek(content):
    return [obj for obj in content if obj['is_current']]
```

```python
get_current_gameweek(json_all['events'])
```

```python
def get_previous_gameweek(content):
    return [obj for obj in content if obj['is_previous']]
```

```python
get_previous_gameweek(json_all['events'])
```
